## Text Preprocessing



### Preprocessing Question/Answer Pairs for Multiple Choice

Each row in the ScienceQA dataset has a 'question' field, which contains question text, and a 'choices' field, which contains a variable length list of multiple choice answers.

BERT models' closest equivalent to answering a multiple choice question is their ability to take in tokenized pair of sentences, divided the by '[SEP]' token, and generate a binary prediction of whether the second follows the first.

So, we have to convert our multiple choice dataset to sentence pairs. We will end up with one question/answer pair for each multiple choice answer in the original ScienceQA dataset.

Note that questions in the ScienceQA dataset have varying number of multiple choice answers. For batch processing, we need a consistent number of answers for each question, so we pad our dataset with empty answers.

````python
MAX_CHOICES = 5 # Max number of choices found in the ScienceQA dataset

def preprocess_function(data):

    # Tokenize question/answer pairs

    first_sentences = []
    second_sentences = []
    tokens = {}

    questions = [ [question] * MAX_CHOICES for question in data['question'] ]
    answers = [ list(choices) + ['']*(MAX_CHOICES-len(choices)) for choices in data["choices"] ]

    # The previous lines return lists of lists, so flatten them
    questions = sum(questions, [])
    answers = sum(answers, [])

    tokenized_examples = tokenizer(questions, answers, truncation=True)

    return {k: [v[i : i + MAX_CHOICES] for i in range(0, len(v), MAX_CHOICES)] for k, v in tokenized_examples.items()}
````

## Image Preprocessing

### Generate Embeddings

To be able to process images as inputs, the model needs to use a vision model. Based on the Huggingface documentation for VisualBert, we will use the Detectron2 library. From this library we will use the MaskRCNN ResNet-101 FPN model which has been pre-trained on the COCO (Common Objects in Context) dataset.

We draw on [this notebook](https://github.com/huggingface/transformers/tree/main/examples/research_projects/visual_bert) provided by HuggingFace, which provides an example of this workflow.

Due to the computationally intensive nature of generating visual embeddings (~1s/image on T4 GPUs, ~2 hours for the dataset), we saved batches of embedding to and S3 AWS bucket, so we can load them as needed.

The code to get the visual embeddings was encapsulated in the GetVisualEmbeddings class, in the visual_embeds.py file.

The detailed models and process used to obtain the embeddings is the following:

1. **Image Preprocessing**: The images are resized, normalized and converted to an ImageList format, required by the model.

2. **ResNET-101 + Feature Pyramid Network Backbone**: This stage extracts multi-scale feature maps from the input images.

3. **Region Proposal Network**: This stage takes the the feature maps and generates region proposals, which are potential bounding boxes that may contain objects of interest. This model generates around 1000 region proposals at different scales and aspect ratios.

4. **Region of Interest Pooling**: The feature maps and region proposals are passed to the ROI pooling layer, which extracts features from each proposed region.

5. **Box Head**: This stage of the process has two tasks:

  5.1. Classification: The box head outputs class scores.

  5.2. Box Regression: The model adjusts the bounding boxes to better fit objects, outputing bounding box deltas.

6. **Box Predictions**: The class scores and bounding box deltas are used to predict the final object bounding boxes and their corresponding confidence scores.

7. **Non-Maximum Suppression (NMS)**: NMS is applied to remove overlapping bounding boxes for the same object.

8. **Box Filterin**: Selected boxes are filtered to ensure a minimum and maximim number of boxes per image.

9. **Visual Embeddings**: For each selected bounding box, the corresponding pooled features from the box head are extracted as visual embeddings.

There were three main challenges when processing the input images to obtain visual embeddings.

1. **Out of Memory**: Balancing memory and execution speed was critical. A batch size of 5 was used because it allowed us to maintain a stable process memory-wise. Moreover, techniques such as handling Python's automatic garbage collection system were used, mainly through the use of ```del``` and ```gc.collect()```.

2. **Processing Time**: Initially, processing the training set visual embeddings was going to take 16 hours on CPU. So we leveraged PyTorch's cuda capabilities to speed up processing. Doing the processing on a GPU was a new challenge in itself.

3. **Dimension Mismatches**: Some batches presented dimension mismatches between what the model expected and the output of previous steps. We were not able to identify the source of this error, and so we opted to discard these faulty batches. The number of lost datapoins is around 500, which does not make it significant.

````python
session = boto3.Session()

s3 = session.client('s3', config=boto3.session.Config(signature_version=botocore.UNSIGNED))

def download_files(bucket_name, local_directory):
    if not os.path.exists(local_directory):
        os.makedirs(local_directory)

    response = s3.list_objects_v2(Bucket=bucket_name)

    while True:
        for item in response.get('Contents', []):
            key = item['Key']
            local_file_path = os.path.join(local_directory, key)

            os.makedirs(os.path.dirname(local_file_path), exist_ok=True)

            s3.download_file(bucket_name, key, local_file_path)
            print(f"Downloaded {key} to {local_file_path}")

        if response['IsTruncated']:
            response = s3.list_objects_v2(Bucket=bucket_name, ContinuationToken=response['NextContinuationToken'])
        else:
            break

bucket_name = 'bertutor'
local_directory = './'

download_files(bucket_name, local_directory)
````

### Add Embeddings to Dataset

Having generating image embeddings and saved them to disk, we need to then load them into our dataset. Some images cause the Detectron library to throw an exception, which we were unable to debug during the time available, so we need to drop any dataset rows for which this occured.

````python
FILEPATH = './visual_embeds/'

def get_visual_embeds_indices(prefix, directory):

  directory_path = os.path.join(FILEPATH, directory)
  files = os.listdir(directory_path)
  indices = []

  for filename in files:
    if filename.startswith(prefix) and filename.endswith(".pt"):
      index_str = filename.split("_")[3].split(".")[0]

      indices.append(int(index_str))

  return indices

def get_visual_embeds_indices_train():
  return get_visual_embeds_indices("visual_embeds_train_", "train")

def get_visual_embeds_indices_test():
  return get_visual_embeds_indices("visual_embeds_test_", "test")

def get_visual_embeds_indices_validation():
  return get_visual_embeds_indices("visual_embeds_validation_", "validation")
````

````python
# Adds 'train' to the visual_embeds_xxxx.pt file name

directory_path = os.path.join(FILEPATH, 'train/')

files = os.listdir(directory_path)

for file in files:
  new_name = 'visual_embeds_train_' + file.split("_")[2]
  old_path = directory_path + file
  new_path = directory_path + new_name
  os.rename(old_path, new_path)
````

````python
batch_size = 5
tokenized_dataset['train'] = tokenized_dataset['train'].add_column('embedding_batch', np.arange(0, len(tokenized_dataset['train'])) // batch_size)
tokenized_dataset['validation'] = tokenized_dataset['validation'].add_column('embedding_batch', np.arange(0, len(tokenized_dataset['validation'])) // batch_size)
tokenized_dataset['test'] = tokenized_dataset['test'].add_column('embedding_batch', np.arange(0, len(tokenized_dataset['test'])) // batch_size)
````

````python
# Filter out rows for which we don't have embeddings

visual_embeds_indices_train = get_visual_embeds_indices_train()
visual_embeds_indices_test = get_visual_embeds_indices_test()
visual_embeds_indices_validation = get_visual_embeds_indices_validation()

def filter_train(row):
  row_batch = row['embedding_batch']
  return row_batch in visual_embeds_indices_train

def filter_validation(row):
  row_batch = row['embedding_batch']
  return row_batch in visual_embeds_indices_validation

def filter_test(row):
  row_batch = row['embedding_batch']
  return row_batch in visual_embeds_indices_test

filtered_tokenized_train = tokenized_dataset['train'].filter(filter_train)
filtered_tokenized_validation = tokenized_dataset['validation'].filter(filter_validation)
filtered_tokenized_test = tokenized_dataset['test'].filter(filter_test)
````

````python
# Load embeddings from disk and into the dataset

def add_visual_embeddings_for_batch(rows, file_prefix, directory):
    batch_idx = rows['embedding_batch'][0]
    for batch_idx_2 in rows['embedding_batch']:
      if batch_idx != batch_idx_2:
        print('Error: misaligned batches')
        return
    filepath = os.path.join(FILEPATH, directory, file_prefix + str(batch_idx) + '.pt')
    loaded_embeds = torch.load(filepath, map_location=torch.device('cpu'))
    visual_embeds = [[tensor.detach().numpy().tolist()] * MAX_CHOICES for tensor in loaded_embeds]
    visual_attention_mask = [[torch.ones(tensor.detach().numpy().shape[:-1], dtype=torch.long).tolist()] * MAX_CHOICES for tensor in loaded_embeds]
    visual_token_type_ids = [[torch.ones(tensor.detach().numpy().shape[:-1], dtype=torch.long).tolist()] * MAX_CHOICES for tensor in loaded_embeds]

    return {'visual_embeds' : visual_embeds, 'visual_attention_mask' : visual_attention_mask, 'visual_token_type_ids' : visual_token_type_ids }

def add_visual_embeddings_for_train_batch(rows):
  return add_visual_embeddings_for_batch(rows, "visual_embeds_train_", "train")

def add_visual_embeddings_for_validation_batch(rows):
  return add_visual_embeddings_for_batch(rows, 'visual_embeds_validation_', "validation")

def add_visual_embeddings_for_test_batch(rows):
  return add_visual_embeddings_for_batch(rows, 'visual_embeds_test_', "test")

tokenized_visual_train = filtered_tokenized_train.map(add_visual_embeddings_for_train_batch, batched=True, batch_size=batch_size)
tokenized_visual_validation = filtered_tokenized_validation.map(add_visual_embeddings_for_validation_batch, batched=True, batch_size=batch_size)
tokenized_visual_test = filtered_tokenized_test.map(add_visual_embeddings_for_test_batch, batched=True, batch_size=batch_size)
````

# Modeling

VisualBERT is a "*simple and flexible model designed for capturing rich semantics in the image and associated text*" [1]

The core idea of VisualBERT is "*to reuse the self-attention mechanism of the Transformer to implicitly align the elements of the input text and regions in the input image*".

In addition to the regular inputs of BERT, VisualBERT also take in visual embeddings, $F$.

Is embedding $F$ is computed by adding three embeddings:

1. A visual feature representation of the bounding box.

2. A segment embedding indicating that it is an image embedding and not a text embedding.

3. A position embedding, that is used when alignments between words and bounding regions is used.

[1]

## Model Training

A complete report of model training is available at: https://api.wandb.ai/links/engai/y1zgp57l

We trained the model using an A100 NVIDIA GPU through the Colab Pro subscription. Here are some important training stats:

 - **Training Time**: 8:16:11

 - **Epochs**: 10

 - **Batch Size**: 16

 - **Initial Learning Rate**: 4e-5

 - **Learning Rate Scheduler**: Linear

 - **Weight Decay**: 0.01

 - **Main Metric**: Accuracy

````python
from dataclasses import dataclass
from transformers.tokenization_utils_base import PreTrainedTokenizerBase, PaddingStrategy
from typing import Optional, Union
import torch

@dataclass
class DataCollatorForMultipleChoice:
    """
    Data collator that will dynamically pad the inputs for multiple choice received.
    """

    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None

    def __call__(self, features):
        label_name = "label" if "label" in features[0].keys() else "labels"
        labels = [feature.pop(label_name) for feature in features]
        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])
        flattened_features = [
            [{k: v[i] for k, v in feature.items()} for i in range(MAX_CHOICES)] for feature in features
        ]
        flattened_features = sum(flattened_features, [])

        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="pt",
        )

        unflattened_batch = {}

        text_embedding_keys = ['input_ids', 'attention_mask', 'token_type_ids']
        for text_embedding_key in text_embedding_keys:
          unflattened_batch[text_embedding_key] = batch[text_embedding_key].view(batch_size, num_choices, -1)

        visual_embedding_keys = ['visual_embeds', 'visual_attention_mask', 'visual_token_type_ids']
        for visual_embedding_key in visual_embedding_keys:
          feature_shape = np.array(features[0][visual_embedding_key]).shape
          unflattened_batch[visual_embedding_key] = batch[visual_embedding_key].view(batch_size, *feature_shape)

        unflattened_batch["labels"] = torch.tensor(labels, dtype=torch.int64)
        return unflattened_batch
````

````python
import evaluate

accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
````

````python
from transformers import VisualBertForMultipleChoice

def load_model_from_wandb(project_name, model_artifact_name):

  run = wandb.init(project=project_name, job_type="model_loading")

  artifact = run.use_artifact(model_artifact_name, type="model")
  artifact_dir = artifact.download()

  model = VisualBertForMultipleChoice.from_pretrained(artifact_dir)
  model.cuda()

  run.finish()

  return model
````

From the model evaluation results, it is clear that the pre-trained VisualBERT model did learn how to perform multiple choice VQA for K-12 science questions.

The not fine-tuned model performs worse than how a random choice model would perform on average, with only a 28% accuracy. # Future Work & Scope of Improvement

Several avenues exist for improving the visual question answering model's performance on the Science QA task. Task-specific pre-training by training masked language modeling with an image objective on the training data could better align the model's representations to the domain.

 We could also dig deeper into the error that caused us to discard hundreds of datapoints, which could boost performance slightly.

  Regarding fine-tuning, evaluating different strategies like updating only the final layers versus tuning the entire image encoder is warranted.
  
  For follow-up exploration, developing a generative model capable of open-ended text generation, akin to GPT rather than BERT's encoder approach, may better handle more complex, open-ended science prompts beyond simple question-answer pairs.
  
  Finally, enhancing multilingual support by complementing the existing English ScienceQA dataset with resources like Hugging Face's multilingual exam dataset could facilitate deploying the system to a wider, global audience of students across different languages and locales.
